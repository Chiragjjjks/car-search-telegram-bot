# simple images file name changed to numbers just for easy reading 

```python
import os

BASE_DIR = r"C:\Users\chira\Downloads\Cars"

IMAGE_EXTENSIONS = {".jpg", ".jpeg", ".png", ".webp"}

for car_folder in os.listdir(BASE_DIR):
    car_path = os.path.join(BASE_DIR, car_folder)

    if not os.path.isdir(car_path):
        continue

    images_dir = os.path.join(car_path, "images")

    if not os.path.isdir(images_dir):
        print(f"Skipping (no images folder): {car_folder}")
        continue

    images = [
        f for f in os.listdir(images_dir)
        if os.path.splitext(f)[1].lower() in IMAGE_EXTENSIONS
    ]

    images.sort()  # ensures consistent ordering

    for idx, filename in enumerate(images, start=1):
        ext = os.path.splitext(filename)[1]
        old_path = os.path.join(images_dir, filename)
        new_path = os.path.join(images_dir, f"{idx}{ext}")

        os.rename(old_path, new_path)

    print(f"Renamed {len(images)} images in {car_folder}")
```

# create a json file with a proper nameing

```python
import os
import json

BASE_DIR = r"C:\Users\chira\Downloads\Cars"

TEMPLATE = {
    "brand": "",
    "model": "",
    "variant": "",
    "year": "",
    "color": "",
    "fuel": "",
    "ownership": "",
    "km": "",
    "price": "",
    "status": "",
    "images": []
}

for folder in os.listdir(BASE_DIR):
    folder_path = os.path.join(BASE_DIR, folder)

    if not os.path.isdir(folder_path):
        continue

    # Check if ANY json file already exists
    json_exists = any(
        file.lower().endswith(".json")
        for file in os.listdir(folder_path)
    )

    if json_exists:
        print(f"Skipping (JSON exists): {folder}")
        continue

    json_path = os.path.join(folder_path, "metadata.json")

    with open(json_path, "w", encoding="utf-8") as f:
        json.dump(TEMPLATE, f, indent=2)

    print(f"Created metadata.json in {folder}")
```
